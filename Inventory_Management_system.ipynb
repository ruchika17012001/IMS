```python
import json
```

```python
f = open("inventory.json",'r')
r = f.read()
f.close()

inventory = json.loads(r)
```

```python
inventory
```

```python
prod_id = str(input("Enter product id:"))
name = str(input("Enter product name:"))
color = str(input("Enter product color:"))
fabric = str(input("Enter product fabric:"))
price = int(input("Enter its price:"))
quantity = int(input("Enter quantity:"))

inventory[prod_id] = {'name': name, 'price': price, 'color' : color, 'fabric' : fabric, 'quantity': quantity}

js = json.dumps(inventory)

f = open("inventory.json",'w')
f.write(js)
f.close()
```
